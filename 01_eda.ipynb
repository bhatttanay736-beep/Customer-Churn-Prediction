```python
import pandas 
import matplotlib.pyplot as plt
import seaborn

a = pandas.read_csv('kaggledataset.csv')

print(a.shape)
print(a.dtypes)
print(a.isnull().sum())
print(a['Churn'].value_counts())

seaborn.countplot(x='Churn', data=a)
plt.title('Churn Distribution')
plt.savefig('churn_distribution.png')
plt.show()

seaborn.countplot(x='Contract', hue='Churn', data=a)
plt.title('Churn by Contract Type')
plt.savefig('churn_by_contract.png')
plt.show()

seaborn.boxplot(x='Churn', y='MonthlyCharges', data=a)
plt.title('Monthly Charges by Churn')
plt.savefig('monthly_charges_churn.png')
plt.show()
```
